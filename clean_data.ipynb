```python
import os
import shutil

# ============ CONFIGURATION ============
SOURCE_DIR = "mrlEyes_2018_01"  # Your dataset folder name
TARGET_DIR = "DataClean"         # Output folder

# ============ CREATE FOLDERS ============
open_dir = os.path.join(TARGET_DIR, "open")
close_dir = os.path.join(TARGET_DIR, "close")
os.makedirs(open_dir, exist_ok=True)
os.makedirs(close_dir, exist_ok=True)

print("🚀 Creating Labeled Dataset")
print("=" * 50)
print(f"Source: {SOURCE_DIR}")
print(f"Target: {TARGET_DIR}")
print("Eye state at position 4: 0=close, 1=open")

# ============ PROCESS ALL FILES ============
open_count = 0
close_count = 0

# Get all subject folders (folders starting with 's')
subjects = [d for d in os.listdir(SOURCE_DIR) 
            if os.path.isdir(os.path.join(SOURCE_DIR, d))]

print(f"\nFound {len(subjects)} subjects")

for subject in subjects:
    subject_path = os.path.join(SOURCE_DIR, subject)
    
    for filename in os.listdir(subject_path):
        if not filename.lower().endswith('.png'):
            continue
        
        # Split filename: s001_00123_0_0_0_0_0_01.png
        parts = filename.split('_')
        
        # Check if we have position 4 (eye state)
        if len(parts) >= 5:
            # Get eye state from position 4 (index 4)
            eye_state = parts[4]
            
            # Remove .png if it's attached
            eye_state = eye_state.replace('.png', '').replace('.PNG', '')
            
            if eye_state == '1':  # Open eye
                dest_folder = open_dir
                open_count += 1
            elif eye_state == '0':  # Close eye
                dest_folder = close_dir
                close_count += 1
            else:
                continue  # Skip if not 0 or 1
            
            # Copy file with unique name
            src = os.path.join(subject_path, filename)
            dst = os.path.join(dest_folder, f"{subject}_{filename}")
            shutil.copy2(src, dst)

# ============ RESULTS ============
print("\n" + "=" * 50)
print("✅ PROCESSING COMPLETE!")
print(f"\n📊 RESULTS:")
print(f"Open eyes: {open_count} images (code '1')")
print(f"Close eyes: {close_count} images (code '0')")
print(f"Total: {open_count + close_count} images")

print(f"\n📁 Folder structure:")
print(f"{TARGET_DIR}/")
print(f"├── open/   ({len(os.listdir(open_dir))} images)")
print(f"└── close/  ({len(os.listdir(close_dir))} images)")

print("\n✨ Dataset ready! You can now use it for ML training.")
```
